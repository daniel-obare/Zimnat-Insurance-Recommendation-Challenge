### Feature Engineering: Train.

```python
products = train[['P5DA', 'RIBP', '8NN1', '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO','BSTQ', 
               'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]
#
trn = train.melt(id_vars=train.columns[:8], value_vars=products, var_name='PCODE', value_name='Label')
trn['ID X PCODE'] = trn['ID'] + ' X ' + trn['PCODE']
trn
```

```python
cat_col = trn[['sex', 'marital_status', 'branch_code', 'occupation_code', 'occupation_category_code', 'PCODE']]
dummy_df = pd.get_dummies(cat_col, dummy_na=True, drop_first=True)
trn = pd.concat([trn, dummy_df], axis=1)
trn.drop(['sex', 'marital_status', 'branch_code', 'occupation_code', 'occupation_category_code', 'PCODE'],
            axis=1, inplace=True)
trn
```

```python
trn['join_day'] = trn['join_date'].apply(lambda x: int(x.split('/')[0]) if (x == x) else np.nan)
trn['join_month'] = trn['join_date'].apply(lambda x: int(x.split('/')[1]) if (x == x) else np.nan)
trn['join_year'] = trn['join_date'].apply(lambda x: int(x.split('/')[2]) if (x == x) else np.nan)
trn['year_diff'] = trn['join_year'] - trn['birth_year']
trn.head()
```

```python
int_col = ['birth_year', 'join_day', 'join_month', 'join_year', 'year_diff']
for col in int_col:
    freq = trn[col].value_counts().to_dict()
    trn[col + '_freq_encode'] = trn[col].map(freq)
#
trn.drop(['join_date', 'ID'], axis=1, inplace=True)
trn
```

### Feature Engineering: Test.

```python
products = test[['P5DA', 'RIBP', '8NN1', '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO','BSTQ', 
               'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3']]
#
tst = test.melt(id_vars=test.columns[:8], value_vars=products, var_name='PCODE', value_name='Label')
tst['ID X PCODE'] = tst['ID'] + ' X ' + tst['PCODE']
#
cat_col = tst[['sex', 'marital_status', 'branch_code', 'occupation_code', 'occupation_category_code', 'PCODE']]
dummy_df = pd.get_dummies(cat_col, dummy_na=True, drop_first=True)
tst = pd.concat([tst, dummy_df], axis=1)
tst.drop(['sex', 'marital_status', 'branch_code', 'occupation_code', 'occupation_category_code', 'PCODE', 'Label'],
            axis=1, inplace=True)
#
tst['join_day'] = tst['join_date'].apply(lambda x: int(x.split('/')[0]) if (x == x) else np.nan)
tst['join_month'] = tst['join_date'].apply(lambda x: int(x.split('/')[1]) if (x == x) else np.nan)
tst['join_year'] = tst['join_date'].apply(lambda x: int(x.split('/')[2]) if (x == x) else np.nan)
tst['year_diff'] = tst['join_year'] - tst['birth_year']
#
int_col = ['birth_year', 'join_day', 'join_month', 'join_year', 'year_diff']
for col in int_col:
    freq = tst[col].value_counts().to_dict()
    tst[col + '_freq_encode'] = tst[col].map(freq)
#
tst.drop(['join_date', 'ID'], axis=1, inplace=True)
tst
```

```python
#export clean data as csv
trn.to_csv('train_clean', index=None, header=True)
tst.to_csv('test_clean', index=None, header=True)
```
